# Data Visualization

## 1. Spiking Heidelberg Digits Dataset (SHD)

```python
import tables
import matplotlib.pyplot as plt
```

```python
def plot_shd_sample_raster():
    
    # Open file
    filename = 'data/shd_train.h5'
    fileh = tables.open_file(filename)
    
    # Rename following variables for convenience
    units = fileh.root.spikes.units
    times = fileh.root.spikes.times
    labels = fileh.root.labels
    
    # Create raster plots
    indicies = (4, 6, 8)
    fig = plt.figure(figsize=(16,4))
    for i, idx in enumerate(indicies):
        ax = plt.subplot(1,3,i+1)
        ax.scatter(times[idx], 700 - units[idx], s=3, alpha=0.7)
        ax.set_xlabel('time (s)')
        ax.set_ylabel('neurons')
        ax.set_title(fileh.root.extra.keys[labels[idx]].decode("utf-8"))
    plt.show()
        
plot_shd_sample_raster()
```
